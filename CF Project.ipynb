```python
import pandas as pd 
```

```python
## we merge the both csv file containing the user rates + the moives 
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'], axis= 1)
ratings.head()
```

```python
## constructing the table which requires user id and the moives title to detemine the ranking for the users and later use it for recommedndation system. 

user_ratings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()

## the output is based on the users x moives
```

```python
## we remove the unnessary moives which have less than 10 users rating it and fill the NAN to 0
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings.head()
```

```python
## we will use the correlation method for the similarity matrix

item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head(50)
```

```python
## build recommendation system

def get_similar_movies(moivename,userrate):
    ## we add 2.5 incase the user provideded a bad rating for the moive. it shows the recommndeation and put other moives in the below
    similar_rate = item_similarity_df[moivename] * (userrate - 2.5)
    similar_rate = similar_rate.sort_values(ascending=False)

    return similar_rate
```

```python
#user = [("2 Fast 2 Furious (Fast and the Furious 2, The) (2003)",5),("The Interview (2014)",0),("12 Years a Slave (2013)", 4)]
user = [("Godfather, The (1972)",5)]
similar_movies = pd.DataFrame()

for movie,ratings in user:
    similar_movies = similar_movies.append(get_similar_movies(movie,ratings),ignore_index=True)

similar_movies.head()
similar_movies.sum().sort_values(ascending=False)
```
